```python
def is_float(n):
    try:
        float(n)
    except:
        return False
    return True
```

```python
df3['total_sqft'].unique()
```

```python
df3[~df3['total_sqft'].apply(is_float)].head(30)
```

```python
def convert_range_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0]) + float(tokens[1]))/2
    try:
        return float(x)
    except:
        return x
```

```python
# converted all other units of area to square feet to increase accuracy
def convert_to_sqft(x):
    if type(x) == float:
        return x
    tokens = x.split('Sq. ')
    if len(tokens) == 2:
        if tokens[1] == 'Meter':
            return float(tokens[0]) * 10.7639
        if tokens[1] == 'Yards':
            return float(tokens[0]) * 9
    if 'Acres' in x:
        return float(x.replace('Acres', '')) * 43560
    if 'Guntha' in x:
        return float(x.replace('Guntha', '')) * 1089    
    if 'Perch' in x:
        return float(x.replace('Perch', '')) * 272.25
```

```python
df4 = df3.copy()
df4['total_sqft'] = df4['total_sqft'].apply(convert_range_to_num)
df4['total_sqft'] = df4['total_sqft'].apply(convert_to_sqft)
```

```python
df4.head()
```

```python
df5 = df4.copy()
df5['price_per_sqft'] = df5['price'] * 100000 / df5['total_sqft']
df5.head()
```

```python
len(df5['location'].unique()) 
```

```python
df5.location = df5['location'].apply(lambda x: x.strip())

location_stats = df5.groupby('location')['location'].agg('count').sort_values(ascending = False)

location_stats_less_than_10 = location_stats[location_stats <= 10]

df5['location'] = df5['location'].apply(lambda x: 'other' if x in location_stats_less_than_10 else x)
df5.head(20)
```

```python
#removing data with per room area < 300 
df6 = df5[~(df5.total_sqft / df5.bhk < 300)]
df6.head(20)
```
